```python
import pandas as pd
df = pd.read_excel('in$fields_сочетания_хоз-в_районов.xlsx', sheet_name = 'Sheet1')
prod_farm_comd_df = pd.read_excel('Вариации продуктов и хоз-в.xlsx', sheet_name = 'Sheet1')

```

```python
prod_cult_df = pd.read_csv('in$product.csv', sep=';', encoding='cp1251')
```

```python
prod_farm_comd_df
```

```python
df1 = df[['farm_nm', 'region_nm']]
df2 = df1.drop_duplicates()
df2 = df2.sort_values(by='farm_nm', ascending=True)
df2
```

```python
final = pd.merge(prod_farm_comd_df, df2, how='inner', left_on = "farm", right_on = 'farm_nm', left_index=False, right_index=False, suffixes=('_x', '_y'))
```

```python
writer = pd.ExcelWriter('Сочетания хоз-в и районов.xlsx')
final.to_excel(writer)
writer.save()
```

```python
final_new = pd.merge(final, prod_cult_df, how='inner', left_on = "product", right_on = 'product_nm', left_index=False, right_index=False)
final_new = final_new[['region_nm', 'crop_nm']]
```
